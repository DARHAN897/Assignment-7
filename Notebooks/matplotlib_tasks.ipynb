```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [19, 21, 22, 20, 23],
    "Score": [85, 78, 90, 88, 92]
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Name"], df["Score"])
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Line Plot")
plt.show()
```

```python
plt.bar(df["Name"], df["Score"])
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")
plt.show()
```
